```python
import os
import pandas as pd

jeju_src = './drive/MyDrive/jeju_card.csv'
jeju_df = pd.read_csv(jeju_src)
jeju_df
```

```python
jeju_df['연도']=jeju_df['연월'].str[:4]
jeju_df.head()
```

```python
jeju_df = jeju_df[jeju_df['연월'].str[:4] != '2016']
jeju_df
```

```python
jeju_df.loc[jeju_df['연령대'] == '20 미만','연령대'] = '20대미만'
jeju_df.head()
```

```python
grouped_df = jeju_df.groupby('연령대').sum()
grouped_df['인당이용금액'] = grouped_df['이용금액'] / grouped_df['이용횟수']
grouped_df = grouped_df.sort_values(by= '인당이용금액',ascending=False)
grouped_df['인당이용금액']
```

```python
grouped_age_sec = jeju_df.groupby(['연령대','업종명'])['이용금액'].sum().reset_index()
grouped_age_sec = grouped_age_sec.sort_values(by=['연령대','이용금액'],ascending=[True,False])

for age in grouped_age_sec['연령대'].unique():
    top3 = grouped_age_sec[grouped_age_sec['연령대'] == age].head(3)['업종명'].tolist()
    print(age,top3)






```
